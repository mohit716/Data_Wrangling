```python
import pandas as pd
# create Dataframe from RewardsData CSV
rewards_df = pd.read_csv("RewardsData.csv")

# view first 20 records
rewards_df.head(20)
                         
```

The Pandas .isna() function is used to identify and return Boolean values (True or False) for each element in a DataFrame or Series, indicating
whether the element is missing or contains a NaN (Not-a-Number) value. True if value is missing and False if the value is present.

```python
rewards_df.isna()
```

```python
# we can then use the .sum() function to get a count of the total number of missing values
missing_value_counts = rewards_df.isna().sum()
missing_value_counts
```

```python
# create and display missing value percentage variable
missing_value_percentages = missing_value_counts / len(rewards_df) * 100
missing_value_percentages
```

The Pandas .fillna function is used to replace any missing (NaN) values in a DataFrame or Series with specified values

```python
rewards_df['Last Seen'].fillna(value=rewards_df['Last Seen'].min(), inplace=True)  #inplace avoids creating separate copy
```

```python
# get a count of total number of missing values
missing_value_counts = rewards_df.isna().sum()
print(missing_value_counts)
```

```python
# drop records with missing values and assign result to new dataframe
rewards_df_sample = rewards_df.dropna()
rewards_df_sample
```

```python
# get a count of tatal number of missing values
missing_value_counts = rewards_df_sample.isna().sum()
print(missing_value_counts)
```

```python
# drop the 'Tags' column from rewards_df_sample
rewards_df_sample = rewards_df_sample.drop(columns='Tags')
rewards_df_sample
```

```python
# get list of unique values for city column
city_unique_values = rewards_df_sample['City'].unique()
# sort values
city_unique_values.sort()
city_unique_values
```

```python
# replace variations of Winston-Salem with the correct format
rewards_df_sample['City'].replace(to_replace=['Winston Salem', 'Winston Salem ',
       'Winston salem', 'Winston salem ', 'Winston-Salem',
       'Winston-Salem ', 'Winston-Salem, NC', 'Winston-salem'], value = "Winston-Salem", inplace=True)
```

```python
# remove all leading and trailing spaces from city column
rewards_df_sample['City'] = rewards_df_sample['City'].str.strip()
```

The str keyword is used here as a way to access methods that oeprate on string objects directly from the class itself, without having to 
create a string instance

```python
# fix any casing issues present in City List
rewards_df_sample['City'] = rewards_df_sample['City'].str.title()
```

```python
# get list of unique values for city column
city_unique_values = rewards_df_sample['City'].unique()
# sort values
city_unique_values.sort()
city_unique_values
```

```python
# get list of unique values for city column
state_unique_values = rewards_df_sample['State'].unique()
# sort values
state_unique_values.sort()
state_unique_values
```

The 'State' column is currently a mixture of state names and abbreviations. A quick web search for "python dicitionary of state of abbreviations" provided a complete dictionary of US state names and abbreviations

```python
# Create a dictionary to map full state names to abbreviations
us_state_to_abbrev = {
    "Alabama": "AL", "Alaska": "AK", "Arizona": "AZ", "Arkansas": "AR", "California": "CA",
    "Colorado": "CO", "Connecticut": "CT", "Delaware": "DE", "Florida": "FL", "Georgia": "GA",
    "Hawaii": "HI", "Idaho": "ID", "Illinois": "IL", "Indiana": "IN", "Iowa": "IA",
    "Kansas": "KS", "Kentucky": "KY", "Louisiana": "LA", "Maine": "ME", "Maryland": "MD",
    "Massachusetts": "MA", "Michigan": "MI", "Minnesota": "MN", "Mississippi": "MS", "Missouri": "MO",
    "Montana": "MT", "Nebraska": "NE", "Nevada": "NV", "New Hampshire": "NH", "New Jersey": "NJ",
    "New Mexico": "NM", "New York": "NY", "North Carolina": "NC", "North Dakota": "ND", "Ohio": "OH",
    "Oklahoma": "OK", "Oregon": "OR", "Pennsylvania": "PA", "Rhode Island": "RI", "South Carolina": "SC",
    "South Dakota": "SD", "Tennessee": "TN", "Texas": "TX", "Utah": "UT", "Vermont": "VT",
    "Virginia": "VA", "Washington": "WA", "West Virginia": "WV", "Wisconsin": "WI", "Wyoming": "WY",
    "District of Columbia": "DC", "Puerto Rico": "PR"
}
```

```python
for state_name in us_state_to_abbrev:
    # Instead of inplace=True, we assign the result back to the column
    rewards_df_sample['State'] = rewards_df_sample['State'].replace(to_replace=state_name, value=us_state_to_abbrev[state_name])

# Verify the changes
print(rewards_df_sample['State'].unique())
```

```python
# Convert us_state_to_abbrev dictionary to DataFrame for merging
states_df = pd.DataFrame(list(us_state_to_abbrev.items()), columns=['State name', 'State'])
states_df.head(3)
```

```python
# Merge the original DataFrame with the states DataFrame
rewards_df_sample=pd.merge(rewards_df_sample, states_df, on='State', how='left')
rewards_df_sample
```

```python
# rename 'State' column to 'State Abbreviation'
rewards_df_sample.rename(columns={'State': 'State Abbreviation'}, inplace=True)
rewards_df_sample.head(3)
```

```python
# 1. Drop the redundant 'State name_x' column
rewards_df_sample.drop(columns=['State name_x'], inplace=True)

# 2. Rename 'State name_y' to just 'State Name'
rewards_df_sample.rename(columns={'State name_y': 'State Name'}, inplace=True)

# 3. View the final clean result
rewards_df_sample.head()
```
